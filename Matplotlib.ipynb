```python
import numpy as np
x=['c','c++','Java','Python']
y=[80,70,75,45]
z=np.array(y)/2
plt.xlabel("langugae")
plt.ylabel("marks")
plt.title("graph")
plt.bar(x,y,label="trend",width=0.5)
plt.bar(x,z,color='r',linewidth=1.1,edgecolor='y')
# withoutlegend does not show the tileinside the graph
plt.legend()
plt.show()
```

```python
x=['c','c++','Java','Python']
y=[80,70,75,45]
z=np.array(y)/2
plt.xlabel("langugae")
plt.ylabel("marks")
plt.title("graph")
plt.bar(x,y,label="trend",width=0.5)
plt.bar([1,2,3,4],z,color='r',linewidth=1.1,edgecolor='y')
# withoutlegend does not show the tileinside the graph
plt.legend()
plt.xticks(x)
plt.show()
```

# PIE CHART

```python
x=[1,2,3,4]
y=['c','c++','java','python']
pt.pie(x,labels=y)
pt.show()
```

```python
# to show the percentage
pt.pie(x,labels=y,autopct="%0.2f")
pt.show()
```

```python
# to out the sliceof the pie char
ex=[0.0,0.0,0.0,0.5]
pt.pie(x,labels=y,autopct="%0.2f",explode=ex)
pt.show()
```

```python
# to define the size of the pie chart
pt.pie(x,labels=y,autopct="%0.2f",radius=0.4)
pt.show()
```

```python
# to get the shadow of pie
pt.pie(x,labels=y,autopct="%0.2f",shadow=True)
pt.show()
```

```python
 # to diatnce mentaine by the 
pt.pie(x,labels=y,autopct="%0.2f",labeldistance=1.5)
pt.show()
```

```python
# to rotate the chart 
pt.pie(x,labels=y,autopct="%0.2f",startangle=180)
pt.show()
```

```python
# change the size of the text
pt.pie(x,labels=y,autopct="%0.2f",textprops={"fontsize":19.5})
pt.show()
```

```python
# to rotate by the clock wise
pt.pie(x,labels=y,autopct="%0.2f",counterclock=False)
pt.show()
```

```python
# to 
pt.pie(x,labels=y, explode=ex, autopct="%0.1f%%", shadow=True, radius=1.3, labeldistance=1.1, startangle=0
,textprops={"fontsize":15},counterclock=False,wedgeprops={"linewidth":5},rotatelabels=False)
```

```python
pt.pie(x,labels=y, explode=ex, autopct="%0.1f%%", shadow=True, radius=1.3, labeldistance=1.1, startangle=0
,textprops={"fontsize":15},counterclock=False,wedgeprops={"linewidth":5},rotatelabels=False)
pt.pie([1],colors='w')
pt.title("piechart")
pt.show()
```

```python
# to create the stem chart
pt.stem(x,y,bottom=0,label="Stem Chart",linefmt=":")
pt.legend()
pt.show()
```

```python
# to create the stem chart
# to color use marker then first letter represent color
pt.stem(x,y,bottom=0,label="Stem Chart",linefmt=":",markerfmt="r*",orientation="horizontal",basefmt="g")
pt.legend()
pt.show()
```

```python
# cReate the area or stack plot
ind=[1,2,3,4,5]
area=[10,20,30,20,10]
```

```python
plt.stackplot(ind,area)
```

```python
area2=[10,20,10,30,70]
plt.stackplot(ind,area,area2,labels=['Area','Area2'])
plt.legend(loc=2)
plt.show()
```

```python
area2=[10,20,10,30,70]
plt.stackplot(ind,area,area2,labels=['Area','Area2'])
plt.legend(loc=2)
plt.title("Stackplot")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

```python
# Create step plot
x=[1,2,3,4,5]
y=[1,2,3,4,5]
plt.step(x,y)
```

```python
plt.step(x,y,color='r',marker="o")
```

```python
plt.step(x,y,color='r',marker="o",ms=30,mfc="k",label="Step plot")
plt.legend()
plt.show()
```

```python
# fill the below cover area by line then use fill between
plt.plot(x,y)
plt.fill_between(x,y)
```

```python
# create the subplot
plt.subplot(2,2,1,title="Sub")
plt.plot(x,y)
plt.subplot(2,2,2)
plt.pie(x,labels=y)
plt.subplot(2,2,3)
plt.boxplot(x)
plt.subplot(2,2,4)
plt.step(x,y)
plt.legend()
plt.show()
```
